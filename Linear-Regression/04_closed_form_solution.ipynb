```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = np.array([0.9, 1.8, 2.4, 3.5, 3.9, 4.4, 5.1, 5.6, 6.3])
S = np.array([1.4, 2.6, 1.0, 3.7, 5.5, 3.2, 3.0, 4.9, 6.3])

W_mean = W.mean()
S_mean = S.mean()
print(W.shape)

print(S.shape)
print("Features (W):\n" , W)
print("Target (S):\n" , S)
```

```python
m = np.sum((W - W_mean) * (S - S_mean)) / np.sum((W - W_mean) ** 2)
c = S_mean - m * W_mean
print("m = ", m)  #Slope
print("c = ", c)  #Intercept
```

```python
S_pred  = m * W + c
print(S_pred)
```

```python
MSE = np.mean((S - S_pred) ** 2)
print("MSE = ", MSE)
R2 = 1 - np.sum((S - S_pred) ** 2) / np.sum((S - np.mean(S)) ** 2)
print("R2 = ", R2)
```

```python
plt.scatter(W, S, color = 'deeppink', label = "Data Points")
plt.plot(W, S_pred , color = 'black', label = "Regression Line")
plt.xlabel("W")
plt.ylabel("S")
plt.legend()
plt.show
```
