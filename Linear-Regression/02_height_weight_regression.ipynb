```python
# IMPORTING LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# DOWNLOADING THE DATASET
dataset = pd.read_csv('/content/weight-height (1).csv')
dataset.head()
```

```python
# DROPPING UNNECESSARY FEATURES
dataset.drop('Gender' , axis = 1)
```

```python
# SPLITTING THE DATASET INTO INPUT AND TARGET VARIABLES
X = dataset['Height'].values.reshape(-1 , 1)    #input variable
Y = dataset['Weight'].values    #target variable
```

```python
# PLOTTING THE DATASET
plt.scatter(X,Y,color = 'deeppink',label = 'Data Points')
plt.legend()
plt.title('Weight vs Height')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# SPLIT DATASET INTO 80% TRAINING AND 20% TESTING
combined  = np.hstack((X,Y.reshape(-1, 1)))
np.random.shuffle(combined)

train_size = int(0.8 * len(combined))
train_data = combined[:train_size]
test_data = combined[train_size:]

X_train = train_data[:, 0].reshape(-1, 1)
Y_train = train_data[:, 1].reshape(-1, 1)

X_test = test_data[:, 0].reshape(-1, 1)
Y_test = test_data[:, 1].reshape(-1, 1)
```

```python
# CALCULATING WEIGHT AND BIAS
mean_x = np.mean(X_train)
mean_y = np.mean(Y_train)

num = np.sum((X_train - mean_x) * (Y_train - mean_y))   # ✅ fixed
den = np.sum((X_train - mean_x) ** 2)

w = num / den
b = mean_y - (w * mean_x)

print("The value of slope is:", w)
print("The value of intercept is:", b)

def predict(x):
    return w * x + b

```

```python
# PLOTTING THE REGRESSION LINE FOR TRAINING SET
plt.scatter(X_train, Y_train, color='deeppink', label='Data Points')
plt.plot(X_train, w*X_train + b, color='black', label='Regression Line')   # ✅ fixed line equation
plt.legend()
plt.title('Weight vs Height')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
#CALCULATING THE MEAN SQUARED ERROR FOR TEST SET
mse = np.mean((Y_train - (w * X_train + b)) ** 2)
print("The mean squared error is:", mse)
```

```python
# PLOTTING THE REGRESSION LINE FOR TEST SET
Y_pred = predict(X_test)
plt.scatter(X_test, Y_test, color='deeppink', label='Data Points')
plt.plot(X_test, w*X_test + b, color='black', label='Regression Line')   # ✅ fixed line equation
plt.legend()
plt.title('Weight vs Height')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
#CALCULATING THE MEAN SQUARED ERROR FOR TRAINING SET
mse = np.mean((Y_test - (w * X_test + b)) ** 2)
print("The mean squared error is:", mse)
```
