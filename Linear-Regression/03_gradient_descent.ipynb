```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = [0.9, 1.8, 2.4, 3.5, 3.9, 4.4, 5.1, 5.6, 6.3]
S = [1.4, 2.6, 1.0, 3.7,  5.5, 3.2, 3.0, 4.9, 6.3]

W = np.array(W)
S = np.array(S)
W_norm = W / np.max(W)
S_norm = S / np.max(S)
print(W.shape)

print(S.shape)
print("Features (W):\n" , W_norm)
print("Target (S):\n" , S_norm)
```

```python
plt.scatter(W_norm, S_norm, color = "deeppink")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Mouse Inspection")
```

```python
# Y = m*X + C
m = 0
c = 0
a = 0.01
iterations = 500
n = len(W_norm)
loss_history = []
m_values = []
c_values = []
```

```python
for i in range(iterations):
  S_pred = m * W_norm + c
  error = S_pred - S_norm

  m_grad = (1/n) * np.sum((error)* W)
  c_grad = (1/n) * np.sum((error))

  m  = m - a * m_grad
  c  = c - a * c_grad

  if i % 50 == 0:
    m_values.append(m)
    c_values.append(c)
  J = np.mean((error) ** 2)
  print("MSE: ", J)
  loss_history.append(J)
```

```python
plt.scatter(W_norm, S_norm, color = "deeppink", label = 'Data Points')
for i in range(len(m_values)):
  plt.plot(W_norm, m_values[i]*W_norm + c_values[i], label = f"Step {i*50}")
plt.title("Gradient Progress")
plt.legend()
plt.show()
```

```python
S_pred_norm = m * W_norm + c
S_pred_original = S_pred_norm * np.max(S)

R_squared_original = 1 - np.sum((S - S_pred_original) ** 2) / np.sum((S - np.mean(S)) ** 2)
print("R-squared (Original Scale):", R_squared_original)
```

```python
plt.scatter(W, S , color = "deeppink", label = 'Data Points')
plt.plot(W, S_pred_original, color = "black", label = 'Regression Line')
plt.title("Mouse Inspection")
plt.xlabel("Weight")
plt.ylabel("Size")
plt.legend()
plt.show()
```
