```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv('/content/weight-height.csv')
dataset.head()
```

```python
dataset.drop('Gender' , axis = 1)
```

```python
X = dataset['Height'].values    #input variable
Y = dataset['Weight'].values    #target variable
```

```python
plt.scatter(X,Y,color = 'blue',label = 'Data Points')
plt.legend()
plt.title('Weight vs Height')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# Split dataset into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Initialize and train the Linear Regression model
model = LinearRegression()

# Reshape X_train and X_test to be 2D arrays
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)

model.fit(X_train, Y_train)
```

```python
# Make predictions
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse_train = mean_squared_error(Y_train, Y_train_pred)
mse_test = mean_squared_error(Y_test, Y_test_pred)
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

print(f"Train RMSE: {rmse_train:.3f}")
print(f"Test RMSE: {rmse_test:.3f}")
print(f"Model coefficient (w): {model.coef_[0]:.3f}")
print(f"Model intercept (b): {model.intercept_:.3f}")
```

```python
# Plot the results
plt.scatter(X_train, Y_train, color='blue', label='Train Data')
plt.scatter(X_test, Y_test, color='green', label='Test Data')
plt.plot(X, model.predict(X.reshape(-1, 1)), color='red', label='Regression Line')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Linear Regression with scikit-learn ')
plt.legend()
plt.show()
```
