```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the dataset and check it
dataset= pd.read_csv('//content/Salary_Data (1).csv')
dataset
```

```python
dataset.shape
```

```python
#Separate the Input Variables & Target Variables
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 1].values
```

```python
plt.scatter(X, Y, color = 'green')
plt.title('Given dataset')
plt.show()
```

```python
# split data into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3, random_state = 1)
```

```python
#Fit the Linear Regression Model to Training set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
#Predict the Test Set Results
Y_pred = lr.predict(X_test)
```

```python
# Check the Model Score on Training & Test sets
print('Train set score',lr.score(X_train,Y_train))
print('Test set score',lr.score(X_test, Y_test))
```

```python
# Visualising the Training set results
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
